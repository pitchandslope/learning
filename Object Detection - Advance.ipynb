# Object Detection

## Good features detection Technique.

### This technique is an extension to Harris corner detection technique. In recent year for corner detection this technique is used insetad of harris detection as it has advance algoritham and we can decide the number of corners we are looking. 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing house image
main = cv2.imread('corner-house.jpg')
main_copy = main.copy()
#converting images from Open Cv compatibility to matplot lib compatibility, so that we can plot them for comparison
main = cv2.cvtColor(main, cv2.COLOR_BGR2RGB)
#just displaying the image
plt.imshow(main)
```

```python
#Converting image values to grayscale for better detection
grayscale_house = cv2.cvtColor(main,cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale_house,cmap='gray')

#Converting image values to float . It is needed for this technique
gray = np.float32(grayscale_house)
plt.imshow(gray,cmap='gray')
```

```python
#Applying good feature formulla
good_feature = cv2.goodFeaturesToTrack(gray, 5, 0.01, 10)
```

```python
#Lets printout the coordinates searched by above formulla
good_feature
```

```python
#As we need to draw the corners on the real image we need to convert coordinates to int
good_feature = np.int0(good_feature)
```

```python
#Drawing circle on the detected corners
for coordinate in good_feature:
    #get value of x, y
    x,y = coordinate.ravel()
    #Draw a circle on coloured image at the x & y location
    cv2.circle(main, (x,y), 3, (255,0,0), -1)

```

```python
#Lets print the image with circles
plt.imshow(main)
```

## From the above output we can see that good_feature can certainly detect one corner that is top-left corner of house.
## But other corners found wasn't usefull.

## Lets try it on a drawing. 


```python
#Importing drawing of the house. 
main = cv2.imread('drawing2.jpg')
main_copy = main.copy()
#converting images from Open Cv compatibility to matplot lib compatibility, so that we can plot them for comparison
main = cv2.cvtColor(main, cv2.COLOR_BGR2RGB)
#just displaying the image
plt.imshow(main)
```

```python
#Converting image values to grayscale for better detection.
#Its already black & white but human eyes and computer algorithams are different, so it is needed.
grayscale_house = cv2.cvtColor(main,cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale_house,cmap='gray')

#Converting image values to float . It is needed for this technique
gray = np.float32(grayscale_house)
plt.imshow(gray,cmap='gray')
```

```python
#Applying good feature formulla to find 10 corners
good_feature = cv2.goodFeaturesToTrack(gray, 10, 0.01, 10)
#As we need to draw the corners on the real image we need to convert coordinates to int
good_feature = np.int0(good_feature)
```

```python
#Drawing circle on the detected corners.
for coordinate in good_feature:
    #get value of x, y
    x,y = coordinate.ravel()
    #Draw a circle on coloured image at the x & y location
    cv2.circle(main, (x,y), 3, (255,0,0), 3)
```

```python
plt.imshow(main)
```
